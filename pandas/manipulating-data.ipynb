## Section Two

```python
# Column from series
seats_columns = pd.Series([5, 5, 5, 5, 5])

# New column called seats
car_sales['Seats'] = seats_columns
car_sales
```

```python
car_sales['Seats'].fillna(5, inplace=True)
```

```python
car_sales
```

```python
# New columns from python list
fuel_economy = [7.5, 9.2, 5.0, 9.6, 8.7, 4.7, 7.6, 8.7, 3.0, 5.6]
car_sales['Fuel per 100KM'] = fuel_economy
car_sales
```

```python
car_sales['Total fuel used (L)'] = car_sales['Odometer (KM)']/100 * car_sales['Fuel per 100KM']
car_sales
```

```python
# Create a column from a single value
car_sales['Number of wheels'] = 4
car_sales
```

```python
car_sales['Passed road safety'] = True
car_sales.dtypes
```

```python
car_sales
```

```python
car_sales.drop('Total fuel used (L)', axis=1, inplace=True)
car_sales
```
