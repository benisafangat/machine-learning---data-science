## 1. An end-to-end Scikit-Learn Workflow

```python
# 1. Get the data ready
import pandas as pd

heart_disease = pd.read_csv('datasets/heart-disease.csv')
heart_disease
```

```python
# Create X (features matrix)
X = heart_disease.drop('target', axis=1)

# Create Y (labels)
y = heart_disease['target']
```

```python
# 2. Choose the right model and hyperparameters
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()

# We'll keep the default hyperparameters
clf.get_params()
```

```python
# 3. Fit the model to the training data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
clf.fit(X_train, y_train)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
# Make a prediction
import numpy as np
```

```python
y_preds = clf.predict(X_test)
```

```python
y_preds
```

```python
y_test
```

```python
# 4. Evaluate the model on the training data and test data
clf.score(X_train, y_train)
```

if it gets 100% on the test data, we might think that someting's wrong

```python
clf.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds)
```

```python
# 5. Improve a model
# Try different amount of n_estimators
np.random.seed(42)
for i in range(10, 100, 10):
    print(f'Trying model with {i} estimators...')
    clf = RandomForestClassifier(n_estimators=i).fit(X_train, y_train)
    print(f'Model accuracy on test set: {clf.score(X_test, y_test) * 100:.2f}%')
    print('')
```

```python
# 6. Save a model and load it
import pickle

pickle.dump(clf, open('models/random_forest_model_1.pkl', 'wb'))
```

```python
loaded_model = pickle.load(open('models/random_forest_model_1.pkl', 'rb'))
loaded_model.score(X_test, y_test)
```
